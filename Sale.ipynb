```python
def creer_table_dates(debut, fin):
    # Créer une plage de dates entre début et fin (en années entières)
    dates = pd.date_range(start=f'{debut}-01-01-2017', end=f'{fin}-12-12-2019', freq='D')
```

```python
df_dates= pd.DataFrame(dates, columns=['dates'])
```
